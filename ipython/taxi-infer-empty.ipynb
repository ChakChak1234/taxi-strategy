```python
import pandas as pd

#use merged data with shortened med/hack, sorted by hack/pickup_datetime, zoned (see csv-to-hdf5.ipynb 
#and -make-zoned-data.ipynb)
taxi_data_file = 'data/taxi_short_zoned_2.csv.gz'

taxi_pd = pd.read_csv(taxi_data_file, compression='gzip', parse_dates=True)
taxi_pd = taxi_pd.drop(['original_index', 'medallion', 'payment_type'], axis=1)
```

```python
#drop garbage columns from csv import/export
taxi_pd = taxi_pd.drop(taxi_pd.columns[:1], axis=1)
```

```python
taxi_pd.columns.values[0] = 'sorted_index'
print taxi_pd.columns
```

```python
min_date = min(taxi_pd['pickup_datetime'])
max_date = max(taxi_pd['dropoff_datetime'])
num_trips = len(taxi_pd)
num_drivers = len(taxi_pd['hack_license'].unique())
total_miles = taxi_pd['trip_distance'].sum()
total_time = taxi_pd['trip_time_in_secs'].sum()
total_total_amount = taxi_pd['total_amount'].sum()
total_fare_amount = taxi_pd['fare_amount'].sum()
total_tip_amount = taxi_pd['tip_amount'].sum()
# total_fees_amount = taxi_pd['surcharge'].sum() + taxi_pd['mta_tax'].sum() + taxi_pd['tolls_amount'].sum()
total_fees_amount = total_total_amount - total_fare_amount - total_tip_amount


print 'Using date range {0} to {1}.'.format(min_date, max_date)
print 'Total of {0:,} trips and {1:,} drivers.'.format(num_trips, num_drivers, grouping=True)
print 'Total of {0:,.2f} miles and {1:,.2f} hours.'.format(total_miles, total_time/3600., grouping=True)
print 'Total of ${0:,.2f} = ${1:,.2f} in fares, ${2:,.2f} in tips, and ${3:,.2f} in tolls/fees.'.format(total_total_amount,
                                                                                                        total_fare_amount,
                                                                                                        total_tip_amount,
                                                                                                        total_fees_amount)
```

```python
taxi_pd['day_of_week'] = taxi_pd.pickup_datetime.apply(lambda x: x.dayofweek)
taxi_pd['pickup_hour'] = taxi_pd.pickup_datetime.apply(lambda x: x.hour)
```

```python
#use euclidean zone-to-zone distances (-make-dist-matrix.ipynb), maybe use zone to zone distances from routing API later
dist_matrix_pd = pd.read_pickle('data/euclid_distance_251x251_pd.pkl')
```

```python
#use euclidean distances to get distance from previous dropoff to current pickup zones
#(we use this as proxy for distance traveled when unoccupied)
#do this simply by renaming dropoff_zone column in dist_matrix_pd and merging
dist_matrix_pd.columns.values[1] = 'prev_dropoff_zone'
taxi_pd = pd.merge(taxi_pd, dist_matrix_pd)
taxi_pd.rename(columns={'euclid_distance' : 'euclid_dist_from_prev_zone'}, inplace=True)
taxi_pd = taxi_pd.sort('sorted_index').drop('prev_dropoff_zone', axis=1)
```
