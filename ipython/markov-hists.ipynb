```python
def transform(rewards):
    mean_reward = mean(rewards)
    zero_scale_rewards = (rewards - mean_reward)*sqrt(2.65/1.56)
    return zero_scale_rewards + mean_reward*2.65/1.56
```

```python
starting_zone_time = 10914
num_shifts = 500
discount = 0.99

from itertools import cycle
ls = ['dotted', 'dashed', 'solid']
ls_cycler = cycle(ls)
lc = ['b', 'b', 'b']
lc_cycler = cycle(lc)
lw = [2, 2, 4]
lw_cycler = cycle(lw)

figure(figsize(10, 6))
rcParams.update({'font.size': 14})

total_rewards_real = loadtxt('figures/mdp/true-hack_performance-weekday-7AM-7PM-rewards.txt')[:num_shifts]
pickups_real = loadtxt('figures/mdp/true-hack_performance-weekday-7AM-7PM-pickups.txt')[:num_shifts]
axvline(mean(total_rewards_real), 0, 100, color='g', lw=2, alpha=0.4)
plot(mean(total_rewards_real), 200, 'g', lw=3, label='real', alpha=0.4)
hist(total_rewards_real, bins=25, color='g', histtype='step', lw=3, fill=True, facecolor='g', alpha=0.35, range=(20,50))

total_rewards = loadtxt('figures/mdp/hack_performance-32-0-0.99-500-10914-0-0-rewards.txt')
total_rewards = transform(total_rewards)
axvline(mean(total_rewards), 0, 100, color='grey', lw=2)
plot(mean(total_rewards), 200, 'grey', lw=3, label='random')
hist(total_rewards, bins=25, color='grey', histtype='step', lw=3, fill=False, range=(20,50))

ylim([0,80])
xlabel('net dollars per hour')
ylabel('number of hacks')
for root_width in [32]:
    for action_width in [32]:
        for depth in [1, 2, 4]:
            s = next(ls_cycler)
            c = next(lc_cycler)
            w = next(lw_cycler)
            total_rewards = loadtxt('figures/mdp/hack_performance-' + str(root_width) + '-'  + str(action_width) + '-'  + 
                                    str(discount) + '-' + str(num_shifts) + '-' + str(starting_zone_time) + '-' + 
                                    str(depth) + '-1-rewards.txt')
            total_rewards = transform(total_rewards)
            axvline(mean(total_rewards), 0, 100, color=c, ls=s, lw=w, alpha=0.4)
            plot(mean(total_rewards), 200, color=c, ls=s, lw=w, label='width = 32, depth = ' + str(depth), alpha=0.4)
            hist(total_rewards, histtype='step', color=c, ls=s, lw=w, fill=(depth==4), facecolor=c,
                 alpha=0.3, range=(20,50), bins=25)
legend(loc=1, fontsize=14)
tight_layout()
savefig('figures/mdp-width-32.png')
show()
```
