```python
import pandas as pd

#use trip data to define zones
trip_data_file = 'data/trip_data_2.h5'

raw_taxi_pd = pd.read_hdf(trip_data_file, 'data')
```

```python
#define lng/lat boundaries

#MSG
center_lat = 40.751
center_lng = -73.992
#Barclays
# center_lat = 40.6826
# center_lng = -73.9747
dlat = 0.01
dlng = 0.01
min_lat = center_lat - dlat
max_lat = center_lat + dlat
min_lng = center_lng - dlng
max_lng = center_lng + dlng

tag = 'MSG-all'
# tag = 'Barclays-all'
# tag = 'none'
min_time = '2013-02-04 10:00:00'
# max_time = '2013-02-05 02:00:00'
# tag = 'lakers'
# min_time = '2013-02-05 14:00:00'
# max_time = '2013-02-06 02:00:00'
# tag = 'mumford'
# min_time = '2013-02-06 14:00:00'
max_time = '2013-02-07 10:00:00'

pick_pd = pd.DataFrame(raw_taxi_pd[((raw_taxi_pd.pickup_latitude > min_lat) &
                                    (raw_taxi_pd.pickup_latitude < max_lat) &
                                    (raw_taxi_pd.pickup_longitude > min_lng) &
                                    (raw_taxi_pd.pickup_longitude < max_lng) &
                                    (raw_taxi_pd.pickup_datetime > min_time) &
                                    (raw_taxi_pd.pickup_datetime < max_time))])
drop_pd = pd.DataFrame(raw_taxi_pd[((raw_taxi_pd.dropoff_latitude > min_lat) &
                                    (raw_taxi_pd.dropoff_latitude < max_lat) &
                                    (raw_taxi_pd.dropoff_longitude > min_lng) &
                                    (raw_taxi_pd.dropoff_longitude < max_lng) &
                                    (raw_taxi_pd.dropoff_datetime > min_time) &
                                    (raw_taxi_pd.dropoff_datetime < max_time))])

import datetime

def str_datetime_to_int64(str_datetime):
    return datetime64(str_datetime).astype(int64)

pick_pd['pickup_datetime_int64'] = pick_pd.pickup_datetime.apply(str_datetime_to_int64)
drop_pd['dropoff_datetime_int64'] = drop_pd.dropoff_datetime.apply(str_datetime_to_int64)
```

```python
#bin times
num_time_int64_bins = 145

min_time_int64 = str_datetime_to_int64(min_time)
max_time_int64 = str_datetime_to_int64(max_time)
time_int64_bins = arange(min_time_int64, max_time_int64 + 1800, 
                         (max_time_int64 + 1800 - min_time_int64) / float(num_time_int64_bins))

pick_ctime = pd.cut(pick_pd.pickup_datetime_int64.values, time_int64_bins)
drop_ctime = pd.cut(drop_pd.dropoff_datetime_int64.values, time_int64_bins)

#total number of stops (pickup or dropoff) in each time bin
pick_time_binned_cnts = pick_pd.groupby(pick_ctime.codes).count()
drop_time_binned_cnts = drop_pd.groupby(drop_ctime.codes).count()

time_bins = pd.to_datetime(time_int64_bins, unit='s')
```

```python
import random, os

def make_scatter_for_time_slice(min_time_slice_int64, max_time_slice_int64, counter, tag):
    if not os.path.exists('figures/scatter-upstream/' + tag):
        os.mkdir('figures/scatter-upstream/' + tag)
    fig = figure(figsize(10, 10))
    rcParams.update({'font.size': 14})
       
    drop_pd_time_slice = pd.DataFrame(drop_pd[(drop_pd.dropoff_datetime_int64 > min_time_slice_int64) &
                                              (drop_pd.dropoff_datetime_int64 < max_time_slice_int64)])
    drop_loc_pd = drop_pd_time_slice[['dropoff_longitude', 'dropoff_latitude']]

    if(len(drop_pd_time_slice) > 0):
        scatter(drop_loc_pd.ix[:,0], drop_loc_pd.ix[:,1], 
                s=10, alpha=1, c='b', lw=0)
        
    pick_pd_time_slice = pd.DataFrame(pick_pd[(pick_pd.pickup_datetime_int64 > min_time_slice_int64) &
                                              (pick_pd.pickup_datetime_int64 < max_time_slice_int64)])
    pick_loc_pd = pick_pd_time_slice[['pickup_longitude', 'pickup_latitude']]
    if(len(pick_pd_time_slice) > 0):
        scatter(pick_loc_pd.ix[:,0], pick_loc_pd.ix[:,1], 
                s=10, alpha=1, c='r', lw=0)
        
    xlim(min_lng, max_lng)
    ylim(min_lat, max_lat)
    plot(2*max_lng, 2*min_lng, 'sr', label='pickups', ms=20, mew=0)
    plot(2*max_lng, 2*min_lng, 'sb', label='dropoffs', ms=20, mew=0)
    leg = legend(loc=1, fontsize=16, numpoints=1, frameon=True)
    leg_frame = leg.get_frame()
    leg_frame.set_facecolor('w')
    leg_frame.set_edgecolor('w')
    gca().get_xaxis().get_major_formatter().set_useOffset(False)
    gca().get_yaxis().get_major_formatter().set_useOffset(False)
    xlabel('longitude (degrees)', size=14)
    ylabel('latitude (degrees)', size=14)
    if tag[:8] == 'Barclays':
        title('Barclays Center')
    if tag[:3] == 'MSG':
        title('Madison Square Garden/Penn Station')
#     close()
#     return fig
    if counter == -1:
        savefig('figures/scatter-upstream/' + tag + '-all.png')
        show()
    else:
        savefig('figures/scatter-upstream/' + tag + '/' + '{0:0>3}'.format(counter) + '.png')
    close()
```
