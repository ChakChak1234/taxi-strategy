```python
import pandas as pd

#use trip data to define zones
trip_data_file = 'data/trip_data_2.h5'

raw_taxi_pd = pd.read_hdf(trip_data_file, 'data')
```

```python
# print raw_taxi_pd.columns
```

```python
#define lng/lat boundaries
center_lat = 40.771
center_lng = -73.87
dlat = 0.005
dlng = 0.01
min_lat = center_lat - dlat
max_lat = center_lat + dlat
min_lng = center_lng - dlng
max_lng = center_lng + dlng

taxi_pd = pd.DataFrame(raw_taxi_pd[((raw_taxi_pd.pickup_latitude > min_lat) &
                                    (raw_taxi_pd.pickup_latitude < max_lat) &
                                    (raw_taxi_pd.pickup_longitude > min_lng) &
                                    (raw_taxi_pd.pickup_longitude < max_lng)) | 
                                   ((raw_taxi_pd.dropoff_latitude > min_lat) &
                                    (raw_taxi_pd.dropoff_latitude < max_lat) &
                                    (raw_taxi_pd.dropoff_longitude > min_lng) &
                                    (raw_taxi_pd.dropoff_longitude < max_lng))])
```

```python
min_date = min(taxi_pd['pickup_datetime'])
max_date = max(taxi_pd['dropoff_datetime'])
num_trips = len(taxi_pd)
num_drivers = len(taxi_pd['hack_license'].unique())
total_miles = taxi_pd['trip_distance'].sum()
total_time = taxi_pd['trip_time_in_secs'].sum()
# total_total_amount = taxi_pd['total_amount'].sum()
# total_fare_amount = taxi_pd['fare_amount'].sum()
# total_tip_amount = taxi_pd['tip_amount'].sum()
# # total_fees_amount = taxi_pd['surcharge'].sum() + taxi_pd['mta_tax'].sum() + taxi_pd['tolls_amount'].sum()
# total_fees_amount = total_total_amount - total_fare_amount - total_tip_amount


print 'Using date range {0} to {1}.'.format(min_date, max_date)
print 'Total of {0:,} trips and {1:,} drivers.'.format(num_trips, num_drivers, grouping=True)
print 'Total of {0:,.2f} miles and {1:,.2f} hours.'.format(total_miles, total_time/3600., grouping=True)
# print 'Total of ${0:,.2f} = ${1:,.2f} in fares, ${2:,.2f} in tips, and ${3:,.2f} in tolls/fees.'.format(total_total_amount,
#                                                                                                         total_fare_amount,
#                                                                                                         total_tip_amount,
#                                                                                                         total_fees_amount)
```

```python
pick_loc_pd = taxi_pd[['pickup_longitude', 'pickup_latitude']]
drop_loc_pd = taxi_pd[['dropoff_longitude', 'dropoff_latitude']]
```

```python
import random

#scatter plot of pickups (red) and dropoffs(blue)
num_scatter_pts = num_trips
figure(figsize(10, 5))
rcParams.update({'font.size': 14})
# axis('off')
scatter_ix = random.sample(range(len(taxi_pd)), num_scatter_pts)
scatter(drop_loc_pd.ix[scatter_ix,0], drop_loc_pd.ix[scatter_ix,1], 
        s=1, alpha=1, c='b', lw=0)
scatter(pick_loc_pd.ix[scatter_ix,0], pick_loc_pd.ix[scatter_ix,1], 
        s=1, alpha=1, c='r', lw=0)
xlim(min_lng, max_lng)
ylim(min_lat, max_lat)
plot(2*max_lng, 2*min_lng, 'sr', label='pickups', ms=20, mew=0)
plot(2*max_lng, 2*min_lng, 'sb', label='dropoffs', ms=20, mew=0)
legend(loc=1, fontsize=16, numpoints=1, frameon=False)
gca().get_xaxis().get_major_formatter().set_useOffset(False)
gca().get_yaxis().get_major_formatter().set_useOffset(False)
xlabel('longitude (degrees)', size=14)
ylabel('latitude (degrees)', size=14)
title('LaGuardia Airport')
tight_layout()
savefig('figures/scatter-LGA.png')
show()
```
