```python
import pandas as pd

#load trip data (med/hack shortened, sorted by hack/pickup date) (see csv-to-hdf5.ipynb)
taxi_file = 'data/taxi_short_2.csv.gz'

taxi_pd = pd.read_csv(taxi_file, compression='gzip')
```

```python
#load zone_matrix from -find-zones.ipynb
zone_matrix = load('data/zone_matrix_40x40.dat')
```

```python
pd.DataFrame(zone_matrix[::-1,:])
```

```python
zones = unique(zone_matrix)
num_zones = len(zones)
```

```python
#define lng/lat boundaries and bins
center_lat = 40.75
center_lng = -73.925
dlat = 0.10
dlng = 0.10
min_lat = center_lat - dlat
max_lat = center_lat + dlat
min_lng = center_lng - dlng
max_lng = center_lng + dlng

num_lat_bins = 40
num_lng_bins = 40

lat_bins = linspace(min_lat, max_lat, num_lat_bins+1)
lng_bins = linspace(min_lng, max_lng, num_lng_bins+1)
lat_lng_mi = pd.MultiIndex.from_product([range(0, num_lat_bins), range(0, num_lng_bins)], names=['lat', 'lng'])
```

```python
zone_dict = append([0],zone_matrix.flatten())
```

```python
def zone_from_lng_lat(lng, lat, lng_bins, lat_bins):
    num_lng_bins = len(lng_bins) - 1
    num_lat_bins = len(lat_bins) - 1
    
    in_range = (min_lng <= lng) * (lng <= max_lng) * (min_lat <= lat) * (lat <= max_lat)
    return zone_dict[vectorize(int)(in_range) * (digitize(lng, lng_bins) + (digitize(lat, lat_bins) - 1)*num_lng_bins)]
```

```python
#convert lng/lat to zone
taxi_zoned_pd = taxi_pd
taxi_zoned_pd['pickup_zone'] = zone_from_lng_lat(taxi_pd.pickup_longitude, taxi_pd.pickup_latitude, lng_bins, lat_bins)
taxi_zoned_pd['dropoff_zone'] = zone_from_lng_lat(taxi_pd.dropoff_longitude, taxi_pd.dropoff_latitude, lng_bins, lat_bins)
```
