```python
import pandas as pd

#use trip data to define zones
trip_data_file = 'data/trip_data_2.h5'

taxi_pd = pd.read_hdf(trip_data_file, 'data')
```

```python
#define lng/lat boundaries
center_lat = 40.75
center_lng = -73.925
dlat = 0.1
dlng = 0.1
min_lat = center_lat - dlat
max_lat = center_lat + dlat
min_lng = center_lng - dlng
max_lng = center_lng + dlng
```

```python
num_lat_bins = 40
num_lng_bins = 40

lat_bins = linspace(min_lat, max_lat, num_lat_bins+1)
lng_bins = linspace(min_lng, max_lng, num_lng_bins+1)

lat_lng_mi = pd.MultiIndex.from_product([range(0, num_lat_bins), range(0, num_lng_bins)], names=['lat', 'lng'])
```

```python
del taxi_pd
```

```python
pick_loc_pd = taxi_pd[['pickup_longitude', 'pickup_latitude']]
drop_loc_pd = taxi_pd[['dropoff_longitude', 'dropoff_latitude']]
```

```python
len_pick = len(pick_loc_pd)
len_drop = len(drop_loc_pd)
num_time_bins = 1

def zone_cnts(time_bin):
    pick_slice = pick_loc_pd[time_bin*(len_pick/num_time_bins):(time_bin+1)*(len_pick/num_time_bins)]
    drop_slice = drop_loc_pd[time_bin*(len_drop/num_time_bins):(time_bin+1)*(len_drop/num_time_bins)]
    pick_clat = pd.cut(pick_slice.pickup_latitude.values, lat_bins)
    pick_clng = pd.cut(pick_slice.pickup_longitude.values, lng_bins)
    drop_clat = pd.cut(drop_slice.dropoff_latitude.values, lat_bins)
    drop_clng = pd.cut(drop_slice.dropoff_longitude.values, lng_bins)

    pick_s = pd.Series(pick_slice.pickup_longitude)
    drop_s = pd.Series(drop_slice.dropoff_longitude)

    #total number of stops (pickup or dropoff) in each lng/lat bin
    stop_binned_cnts = pick_s.groupby([pick_clat.codes, pick_clng.codes]).count() + \
                       drop_s.groupby([drop_clat.codes, drop_clng.codes]).count()
    stop_binned_cnts_ri = stop_binned_cnts.reindex(lat_lng_mi).fillna(0)
    
    #define zones = top #=num_zones - 1 bins with the most stops
    #lump all other stops in other bins (that were previously nonzero) into zone_other
    num_zones = 250
    zone_cutoff_cnt = sort(stop_binned_cnts_ri)[-num_zones]
    zone_cnts = stop_binned_cnts_ri[stop_binned_cnts_ri > zone_cutoff_cnt]
    zone_zero_cnts = stop_binned_cnts_ri[stop_binned_cnts_ri < 10.]
    zone_other_cnts = sum(stop_binned_cnts_ri[stop_binned_cnts_ri <= zone_cutoff_cnt])

    zone_cnts_ri = zone_cnts.reindex(lat_lng_mi).fillna(float(zone_other_cnts) / (num_lat_bins*num_lng_bins - num_zones - len(zone_zero_cnts) + 1))
    zone_cnts_ri[stop_binned_cnts_ri < 10.] = 0.
    return reshape(zone_cnts_ri.values, (num_lat_bins, num_lng_bins))
```

```python
from matplotlib.patches import FancyArrowPatch
from mpl_toolkits.mplot3d import proj3d

class Arrow3D(FancyArrowPatch):
    def __init__(self, xs, ys, zs, *args, **kwargs):
        FancyArrowPatch.__init__(self, (0,0), (0,0), *args, **kwargs)
        self._verts3d = xs, ys, zs

    def draw(self, renderer):
        xs3d, ys3d, zs3d = self._verts3d
        xs, ys, zs = proj3d.proj_transform(xs3d, ys3d, zs3d, renderer.M)
        self.set_positions((xs[0],ys[0]),(xs[1],ys[1]))
        FancyArrowPatch.draw(self, renderer)
```
