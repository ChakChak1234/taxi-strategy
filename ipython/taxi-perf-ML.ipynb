```python
import pandas as pd

#use merged data with shortened med/hack, sorted by hack/pickup_datetime, zoned, empty behavior inferred
#(see csv-to-hdf5.ipynb, -make-zoned-data.ipynb, -infer-empty.ipynb)
taxi_data_file = 'data/taxi_short_zoned_empty_2.csv.gz'

taxi_pd = pd.read_csv(taxi_data_file, compression='gzip', parse_dates=True)

#drop garbage columns from csv import/export
taxi_pd = taxi_pd.drop(taxi_pd.columns[:1], axis=1)
```

```python
#make some basic cuts
taxi_pd = pd.DataFrame(taxi_pd[(taxi_pd.trip_distance > 0) &
                               (taxi_pd.trip_time_in_secs > 0) &
                               (taxi_pd.time_since_prev_fare > 0) ])
```

```python
#take only drivers with over x trips
taxi_pd = taxi_pd.groupby('hack_license').filter(lambda x: len(x) > 300)
```

```python
#calculate some performance stats for each driver
perf_pd = pd.DataFrame([])
perf_pd['hack_license'] = taxi_pd.hack_license.unique()
perf_pd['total_time_on_shift'] = taxi_pd.groupby('hack_license').trip_time_in_secs.sum().values + \
                                 taxi_pd.groupby('hack_license').time_since_prev_fare.sum().values
perf_pd['total_dist_on_shift'] = taxi_pd.groupby('hack_license').trip_distance.sum().values + \
                                 taxi_pd.groupby('hack_license').euclid_dist_from_prev_zone.sum().values
perf_pd['avg_speed_mph'] = perf_pd.total_dist_on_shift / (perf_pd.total_time_on_shift / 3600.)
perf_pd['total_fare'] = taxi_pd.groupby('hack_license').fare_amount.sum().values
perf_pd['total_tip'] = taxi_pd.groupby('hack_license').tip_amount.sum().values
perf_pd['total_total'] = taxi_pd.groupby('hack_license').total_amount.sum().values
perf_pd['fare_per_hour'] = perf_pd.total_fare / (perf_pd.total_time_on_shift / 3600.)
perf_pd['tip_per_hour'] = perf_pd.total_tip / (perf_pd.total_time_on_shift / 3600.)
perf_pd['dollars_per_hour'] = perf_pd.fare_per_hour + perf_pd.tip_per_hour
perf_pd['dollars_per_mile'] = (perf_pd.total_fare + perf_pd.total_tip) / perf_pd.total_dist_on_shift

miles_per_gallon = 15.
dollars_per_gallon = 3.
dollars_per_mile = dollars_per_gallon / miles_per_gallon

perf_pd['net_dollars_per_hour'] = perf_pd.dollars_per_hour - dollars_per_mile*perf_pd.avg_speed_mph

net_dollars_per_hour_bot = percentile(perf_pd.net_dollars_per_hour, 5)
net_dollars_per_hour_top = percentile(perf_pd.net_dollars_per_hour, 95)

def performance(net_dollars_per_hour):
    return 1 * (net_dollars_per_hour >= net_dollars_per_hour_top) + \
          -1 * (net_dollars_per_hour <= net_dollars_per_hour_bot)

perf_pd['performance'] = perf_pd.net_dollars_per_hour.apply(performance)
```

```python
#define lng/lat boundaries
center_lat = 40.75
center_lng = -73.925
dlat = 0.1
dlng = 0.1
min_lat = center_lat - dlat
max_lat = center_lat + dlat
min_lng = center_lng - dlng
max_lng = center_lng + dlng
num_lat_bins = 40
num_lng_bins = 40
lat_bins = linspace(min_lat, max_lat, num_lat_bins+1)
lng_bins = linspace(min_lng, max_lng, num_lng_bins+1)
```

```python
def zone_to_lat_lng(zone):
    zone_lat_bin = floor(zone / num_lng_bins) + 1
    zone_lng_bin = zone % num_lng_bins
    
    lng = min_lng + dlng/num_lng_bins + zone_lng_bin*2.*dlng/num_lng_bins
    lat = min_lat + dlat/num_lat_bins + zone_lat_bin*2.*dlat/num_lat_bins
    return array([lat, lng])
```

```python
def make_zone_weight_map(day_of_week, time_bin):
    vminmax = 0.15
    dv = 0.05
#     vminmax = 0.05
#     dv = 0.01
    clf()
    figsize(10,10)
    zone_weight_slice = pickup_zone_time_weight_pd[(pickup_zone_time_weight_pd.day_of_week == day_of_week) &
                                                   (pickup_zone_time_weight_pd.time_bin == time_bin)][['zone', 'weight']]
    zones = zone_weight_slice.zone.unique()
    num_zones = len(zones)
    zone_weight_map = -ones(num_lng_bins*num_lat_bins)
    zone_weight_map[zones[1:]] = zone_weight_slice.weight.values
    zone_weight_map[zone_weight_map == -1.] = zone_weight_slice.weight.values[0]
    extent = [lng_bins[0], lng_bins[-1], lat_bins[0], lat_bins[-1]]
    zone_weight_map_reshape = zone_weight_map.reshape((num_lat_bins, num_lng_bins))
    imshow(zone_weight_map_reshape, interpolation='none', origin='lower', 
           extent=extent, vmin=-vminmax, vmax=vminmax)
    xlabel('longitude (degrees)', size=18)
    ylabel('latitude (degrees)', size=18)
    cbar = colorbar(shrink=0.7, ticks=arange(-vminmax,vminmax+dv,dv), pad=0.025)
    cbar.set_label('weight', rotation=270, labelpad=13, size=18)
    cbar.set_ticklabels([r'$\leq$'+str(vminmax)] + \
                        ['{0:.2f}'.format(x) for x in arange(-vminmax+dv,vminmax,dv)] + \
                        [r'$\geq$'+str(vminmax)])
    if day_of_week == 1:
        day_of_week_tag = 'Weekday'
    else:
        day_of_week_tag = 'Weekend'
    time_tag = '{0:0>2}'.format(time_bin*2) + ':00 - ' + '{0:0>2}'.format(mod((time_bin+1)*2, 24)) + ':00'
    title(day_of_week_tag + ': ' + time_tag, fontsize=24)
#     savefig('figures/zone-weight-L1/' + day_of_week_tag + '-' + '{0:0>2}'.format(time_bin*2) + '-' + \
#             '{0:0>2}'.format(mod((time_bin+1)*2, 24)))
    tight_layout()
    savefig('figures/zone-weight-L1/' + day_of_week_tag + '-' + '{0:0>2}'.format(time_bin))
    savetxt('figures/zone-weight-L1/data/' + day_of_week_tag + '-' + '{0:0>2}.txt'.format(time_bin), zone_weight_map_reshape)
#     savefig('figures/zone-weight/' + day_of_week_tag + '-' + '{0:0>2}'.format(time_bin))
#     savetxt('figures/zone-weight/data/' + day_of_week_tag + '-' + '{0:0>2}.txt'.format(time_bin), zone_weight_map_reshape)
    close()
#     show()
```

```python
for day_of_week in [1,2]:
    for time_bin in range(12):
        make_zone_weight_map(day_of_week, time_bin)
```
